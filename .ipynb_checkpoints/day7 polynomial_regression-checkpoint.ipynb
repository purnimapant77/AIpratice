```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
```

```python
X=8*np.random.rand(250,1)-4 #generating data randomly
y=0.8+X**2+0.6*X+2+np.random.randn(250,1)
```

```python
plt.plot(X,y,"b.")
plt.xlabel("X")
plt.ylabel("y")
plt.title("representation of random data")
plt.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X.shape
```

```python
y.shape
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.plot(X,y,"b.")
plt.show()
```

```python
#since linear regression doesnt give good oor required output for this dataset
#now applying polynomial linear regression
#p2=PolynomialFeatures(degree=2,include_bias=False) # since degree is it it make 3 columns from degree 0 to 2
p2=PolynomialFeatures(degree=2)
X_train_trans=p2.fit_transform(X_train)  #include_bias=False remove x^0 term 
X_test_trans=p2.transform(X_test)
```

```python
X_train[0]
```

```python
X_train_trans[0]
```

```python
#we apply polynomial feature onboth train and test input data but dont need in output data
lr=LinearRegression()
lr.fit(X_train_trans,y_train)
```

```python
y_pred=lr.predict(X_test_trans)
```

```python
r2_score(y_test,y_pred)
```

```python
# after applying this it give too much improvement in model
#now using polnomial of more degree
p5=PolynomialFeatures(degree=5)
X_train_5=p5.fit_transform(X_train)
X_test_5=p5.transform(X_test)
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train_5,y_train)
y_pred=lr.predict(X_test_5)
r2_score(y_test,y_pred)
```
