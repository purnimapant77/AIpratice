```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import load_wine
```

```python
data=load_wine()
data.keys()
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X=df[data.feature_names]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
X_scaled=ss.fit_transform(X)
```

```python
km=KMeans(n_clusters=4, random_state=42,n_init=10) #n_init=10 mean try for 10 time
```

```python
df['Group']=km.fit_predict(X)
```

```python
plt.figure(figsize=(6,5))
for group in df['Group'].unique():
    group_data=df[df['Group']==group]
    plt.scatter(group_data[data.feature_names[0]],
                group_data[data.feature_names[1]],
                label=f"Cluster {group}")
```
